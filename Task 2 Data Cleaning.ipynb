```python
def remove_extra(st):
    for i in st:
        if i == "(" or i == '"':
            index = st.index(i)
            return st[0:index]
    return st
```

```python
remove_extra('this is "erger" hj')
```

```python


data["name"] = data["name"].apply(remove_extra)

```
